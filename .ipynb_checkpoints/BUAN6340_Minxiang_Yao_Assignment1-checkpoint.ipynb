#### 3)	How many players have hit 40 or more HRs in one single season? (Number only)

#### 10)	Top 5 highest OBP (on base percentage) with at least 500 PAs in 1977?  (Dataframe)

```python
df1977 = df[(df['yearID']==1977)]
df['OBP'] = (df['H'] + df['BB'] + df['IBB'] + df['SH'] + df['SF'])/df['AB']
df['PA'] = df['AB'] + df['BB'] + df['IBB'] + df['SH'] + df['SF']
```

```python
df1977.groupby('OBP')
df1 = df1977[(df1977['PA']>=500)]
df1.sort_values(by='OBP',ascending=False).head(5)
```

#### 12)	Leaders in hits from 1940 up to and including 1949. (Dataframe)

```python
df12 = df[(df['yearID']>=1940) & (df['yearID']<=1949)].groupby(['playerID','nameFirst','nameLast','yearID'],as_index = False).agg({'H':sum})
df12max = df12.groupby('yearID').apply(lambda y_max:y_max[y_max.H==y_max.H.max()])
df12max
```

#### 13)	Who led MLB with the most hits the most times?  And how many times?  (Dataframe, Number)name column number of time（years）

```python
df13= df.groupby(['yearID','playerID','nameFirst','nameLast']).agg({'H':sum})
df13.groupby('yearID').agg({'H':max})
df13m = df13.max(level=0).reset_index().merge(df13.reset_index())
df13m.groupby(['playerID','nameFirst','nameLast']).size().reset_index(name='counts').max()['counts']
dfm = df13m.groupby(['playerID','nameFirst','nameLast']).size().reset_index(name='counts')
dfplayer = dfm[(dfm['counts']==7)]
df13m.merge(dfplayer.reset_index())
```
