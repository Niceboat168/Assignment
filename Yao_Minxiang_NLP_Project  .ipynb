```python
df['label']=df['label'].astype(float)
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df.label 
df.drop("label", axis=1) 
X_train, X_test, y_train, y_test = train_test_split(df['text'], y)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
y=df.label
doc_terms_train, doc_term_test, y_train, y_test=train_test_split(df['text'],y)
```

```python
count_vec=TfidfVectorizer(binary=False, decode_error='ignore', stop_words='english')
```

```python
X_train= count_vec.fit_transform(doc_terms_train)
X_test = count_vec.transform(doc_term_test)
x = count_vec.transform(df['text'])
print(doc_terms_train)
```

```python
print(count_vec.get_feature_names())  
print(X_train.toarray())  
print(y)
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
print ("Classification rate for NB:", model.score(X_test, y_test))
```
